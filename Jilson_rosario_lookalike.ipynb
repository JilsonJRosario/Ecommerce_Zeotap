```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

customers_df = pd.read_csv("C:/Users/hp/Downloads/Customers.csv")
products_df = pd.read_csv("C:/Users\hp/Downloads/Products.csv")
transactions_df = pd.read_csv("C:/Users/hp/Downloads/Transactions.csv")
```

```python
# Preprocessing
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
```

```python
# Merge datasets
transactions_products_df = pd.merge(transactions_df, products_df, on='ProductID', how='left')
complete_data = pd.merge(transactions_products_df, customers_df, on='CustomerID', how='left')
```

```python
# Aggregate customer transaction behavior
customer_features = complete_data.groupby('CustomerID').agg(
    total_spent=('TotalValue', 'sum'),
    total_transactions=('TransactionID', 'nunique'),
    avg_transaction_value=('TotalValue', 'mean'),
    unique_categories=('Category', 'nunique'),
    region=('Region', 'first')
).reset_index()

# Pivot table for product category preferences
category_pivot = pd.pivot_table(
    complete_data, 
    values='Quantity', 
    index='CustomerID', 
    columns='Category', 
    aggfunc='sum', 
    fill_value=0
)

# Combine customer features and product preferences
customer_profiles = pd.merge(customer_features, category_pivot, on='CustomerID', how='left')
```

```python
# Normalize numerical features
scaler = StandardScaler()
numeric_features = customer_profiles.iloc[:, 1:].select_dtypes(include=['float64', 'int64'])
normalized_features = scaler.fit_transform(numeric_features)

# Compute similarity matrix
similarity_matrix = cosine_similarity(normalized_features)
similarity_df = pd.DataFrame(similarity_matrix, index=customer_profiles['CustomerID'], columns=customer_profiles['CustomerID'])
```

```python
# Generate lookalike recommendations for the first 20 customers
lookalikes = {}
for customer_id in customer_profiles['CustomerID'][:20]:
    top_similar = similarity_df[customer_id].nlargest(4)[1:4]  # Exclude self
    lookalikes[customer_id] = list(zip(top_similar.index, top_similar.values))

# Convert lookalikes to DataFrame and save as CSV
lookalike_df = pd.DataFrame.from_dict(lookalikes, orient='index', columns=['SimilarCustomer1', 'SimilarCustomer2', 'SimilarCustomer3'])
lookalike_df.to_csv('C:/Users/hp/Downloads/Lookalike.csv', index_label='CustomerID')

# Output
print("Lookalike recommendations for the first 20 customers have been saved to Lookalike.csv.")
```

The above code generates lookalike recommendations for the first 20 customers based on similarity scores. It goes through each customer and picks top 3 most similar customers by picking the highest similarity values from a similarity matrix excluding the customer itself. It stores these lookalikes in a dictionary, gets the dictionary, and converts it into a DataFrame that gets saved as a CSV file on your computer under the name "Lookalike.csv". Lastly, a success message is echoed to the user that the file is created and saved.

```python
from sklearn.metrics.pairwise import pairwise_distances

# Create user-product matrix
user_product_matrix = complete_data.pivot_table(
    index='CustomerID', 
    columns='ProductID', 
    values='Quantity', 
    aggfunc='sum', 
    fill_value=0
)

# Calculate pairwise user similarity
user_similarity = 1 - pairwise_distances(user_product_matrix, metric='cosine')
user_similarity_df = pd.DataFrame(user_similarity, index=user_product_matrix.index, columns=user_product_matrix.index)

# Generate collaborative filtering recommendations
collaborative_lookalikes = {}
for customer_id in user_product_matrix.index[:20]:
    top_similar = user_similarity_df[customer_id].nlargest(4)[1:4]  # Exclude self
    collaborative_lookalikes[customer_id] = list(zip(top_similar.index, top_similar.values))

# Save collaborative filtering recommendations
collaborative_lookalike_df = pd.DataFrame.from_dict(collaborative_lookalikes, orient='index', columns=['SimilarCustomer1', 'SimilarCustomer2', 'SimilarCustomer3'])
collaborative_lookalike_df.to_csv('C:/Users/hp/Downloads/Collaborative_Lookalike.csv', index_label='CustomerID')
```

This code implements collaborative filtering to generate lookalike recommendations based on customer-product interactions. It starts by building a user-product matrix from the complete dataset, where rows represent customers, and columns represent products, which includes the quantity purchased. Then, it calculates the pairwise cosine similarity between users, which measures how similar customers are based on their purchasing patterns. For the first 20 customers, it finds the top 3 most similar customers excluding the customer itself and stores these lookalikes. The recommendations are saved in a CSV file called "Collaborative_Lookalike.csv" on the user's computer.

There are two recommendation techniques, namely Cosine Similarity and Collaborative Filtering. Both exhibit different focuses and strengths in identifying similarities among customers. Cosine Similarity consistently identifies customers with a high similarity score based on general patterns of behavior or shared characteristics. In comparison, Collaborative Filtering provides recommendations based on the interactions of the users and the items or historical activity. This approach often brings out more divergent suggestions of lower similarity score, emphasizing a unique aspect that customer preferences and the transaction history exhibit and are unable to capture via Cosine Similarity.

Despite their differences, these methods complement each other very well. Cosine Similarity is very good at finding customers with similar features or who behave alike, while Collaborative Filtering reveals subtle relationships through past interactions. Combining these techniques provides a more complete understanding of customer behavior, allowing businesses to identify both broad patterns and subtle connections that may be missed when relying on a single method.
